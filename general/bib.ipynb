```python
from collections import Counter
import re

bib_file = "data_aux/dat_ignore/refs.bib"   # change to your .bib filename

# Extract citekeys from lines like:
# @article{key,
pattern = re.compile(r'@\w+\s*{\s*([^,]+)\s*,')

with open(bib_file, "r", encoding="utf-8") as f:
    content = f.read()

citekeys = pattern.findall(content)

# Count occurrences
counts = Counter(citekeys)

# Print duplicates
duplicates = {k: v for k, v in counts.items() if v > 1}

if duplicates:
    print("Duplicate citekeys found:\n")
    for key, count in duplicates.items():
        print(f"{key}  -->  {count} times")
else:
    print("No duplicate citekeys found.")
```
